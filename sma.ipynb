```python
enc1=tiktoken.encoding_for_model("gpt-4o")
enc2=tiktoken.get_encoding("cl100k_base")
text="Hello"
tokens1=enc1.encode(text)
tokens2=enc2.encode(text)

print("Tokens1:",tokens1)
print("Tokens2:",tokens2)

for t in tokens1:
    print("token1:", t, "->", enc1.decode([t]))
for t in tokens2:
    print("token2:", t, "->", enc2.decode([t]))
```

```python
import csv
import tiktoken

enc = tiktoken.get_encoding("cl100k_base")
vocab = enc._mergeable_ranks

with open("mergeable_ranks.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f, quoting=csv.QUOTE_ALL, escapechar="\\")

    writer.writerow(["token_bytes", "decoded", "rank"])

    for token, rank in vocab.items():
        token_bytes = repr(token)

        try:
            decoded = token.decode("utf-8")
        except UnicodeDecodeError:
            decoded = ""

        decoded = decoded.replace("\n", "\\n").replace("\r", "\\r")

        writer.writerow([token_bytes, decoded, rank])
```
